```python
x=[]
yn=[]
yc = []
yid = []
for j in range(time+1):
    cn0 = df_g['Cn0DbHz']['t'+str(j)]
    svid = df_g['Svid']['t'+str(j)]
    x.append(j)
    yn.append(len(cn0))
    yc.append(cn0.mean())
    yid.append(svid)
```
